# Testes

### Instanciando svm com kernel linear

```python
from sklearn import svm

linear_svm1_400 = svm.SVC(kernel='linear')
linear_svm2_400 = svm.SVC(kernel='linear')
linear_svm3_400 = svm.SVC(kernel='linear')
linear_svm4_400 = svm.SVC(kernel='linear')
linear_svm5_400 = svm.SVC(kernel='linear')
```

```python
linear_result_400 = []
```

```python
linear_svm1_400.fit(x400_train1, y400_train1)
```

```python
linear_score = linear_svm1_400.score(x400_test1, y400_test1)
linear_result_400.append(linear_score)
print(linear_score)
```

```python
linear_svm2_400.fit(x400_train2, y400_train2)
```

```python
linear_score = linear_svm2_400.score(x400_test2, y400_test2)
linear_result_400.append(linear_score)
print(linear_score)
```

```python
linear_svm3_400.fit(x400_train3, y400_train3)
```

```python
linear_score = linear_svm3_400.score(x400_test3, y400_test3)
linear_result_400.append(linear_score)
print(linear_score)
```

```python
linear_svm4_400.fit(x400_train4, y400_train4)
```

```python
linear_score = linear_svm4_400.score(x400_test4, y400_test4)
linear_result_400.append(linear_score)
print(linear_score)
```

```python
linear_svm5_400.fit(x400_train5, y400_train5)
```

```python
linear_score = linear_svm5_400.score(x400_test5, y400_test5)
linear_result_400.append(linear_score)
print(linear_score)
```

```python
linear_result_400 = np.array(linear_result_400)
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (linear_result_400.mean(), linear_result_400.std() * 2))
```

```python
import numpy as np  
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img  
from keras.models import Sequential  
from keras.layers import Dropout, Flatten, Dense  
from keras import applications  
from keras.utils.np_utils import to_categorical  
import matplotlib.pyplot as plt  
import math  
```

```python
#Dimensoes de entrada da VGG16
img_width, img_height = 224, 224  

batch_size = 16

#Path dos dados a serem utilizados
data_dir40 = '../Dados/breast_cancer40'
data_dir100 = '../Dados/breast_cancer100'
data_dir200 = '../Dados/breast_cancer200'
data_dir400 = '../Dados/breast_cancer400'
```

# Carregar as features de magnitude 100x 

```python
datagen = ImageDataGenerator(rescale=1. / 255)
#Carregando fator 100x
generator100 = datagen.flow_from_directory(
    data_dir100,
    batch_size=batch_size,
    target_size=(img_width, img_height),
    class_mode='categorical',
    shuffle=False)

#Features extraidas
data100 = np.load('bottleneck_features_100.npy')
len_Data_100 = len(data100)
data100 = data100.reshape(len_Data_100,-1)
#Labels em ordem dos dados
labels100 = generator100.classes
num_classes100 = len(generator100.class_indices)
```

# Dividir o conjunto de dados em treino e teste com 80% dos dados para treino

```python
from sklearn.model_selection import train_test_split

X100_train, X100_test, y100_train, y100_test = train_test_split(data100, labels100, random_state=0, train_size=0.8)

len_train_100 = len(X100_train)
X100_train = X100_train.reshape(len_train_100,-1)

len_test_100 = len(X100_test)
X100_test = X100_test.reshape(len_test_100,-1)
```
